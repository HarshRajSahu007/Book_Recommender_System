## Popularity Based Recommendation System

```python
books.head()
```

```python
books.isnull().sum()
```

```python
ratings.isnull().sum()
```

```python
ratings_with_name=ratings.merge(books,on='ISBN')
```

```python
num_rating_df=ratings_with_name.groupby("Book-Title").count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
num_rating_df
```

```python
ratings_with_name['Book-Rating'] = pd.to_numeric(ratings_with_name['Book-Rating'], errors='coerce')
```

```python
ratings_with_name = ratings_with_name.dropna(subset=['Book-Rating'])
```

```python
ratings_with_name['Book-Rating'].fillna(ratings_with_name['Book-Rating'].median(), inplace=True)
```

```python
avg_rating_df = ratings_with_name.groupby("Book-Title").mean(numeric_only=True)['Book-Rating'].reset_index()
avg_rating_df.rename(columns={'Book-Rating': 'avg_ratings'}, inplace=True)
print(avg_rating_df)
```

```python
popular_df=num_rating_df.merge(avg_rating_df,on="Book-Title")
popular_df
```

```python
popular_df=popular_df[popular_df["num_ratings"]>=250].sort_values('avg_ratings',ascending=False).head(50).reset_index()
```

```python
popular_df
```

```python
popular_df=popular_df.merge(books,on="Book-Title").drop_duplicates("Book-Title")[['Book-Title','Book-Author','Image-URL-M','num_ratings','avg_ratings']]
```

```python
popular_df
```

## Collaborative Filtering Based Recommendation System

```python
x=ratings_with_name.groupby('User-ID').count()['Book-Rating']>200
good_users=x[x].index
```

```python
filtered_ratings=ratings_with_name[ratings_with_name['User-ID'].isin(good_users)]
```

```python
y=filtered_ratings.groupby('Book-Title').count()['Book-Rating']>=50
famous_books=y[y].index
```

```python
famous_books
```

```python
final_ratings=filtered_ratings[filtered_ratings['Book-Title'].isin(famous_books)]
```

```python
final_ratings
```

```python
final_ratings.drop_duplicates()
```

```python
pt=final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores=cosine_similarity(pt)
```

```python
similarity_scores
```

```python
def recommend(book_name):
    index=np.where(pt.index==book_name)[0][0]
    similar_items=sorted(list(enumerate(similarity_scores[index])),key=lambda x : x[1],reverse=True)[1:6]
    data=[]
    for i in similar_items:
        item=[]
        temp_df=books[books['Book-Title']==pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
        
    return data
    
```

```python
recommend('Year of Wonders')
```
